# 3. Handling missing values

```python
# Convert "Number of employees" to numeric and replace errors with NaN
df["Number of employees"] = pd.to_numeric(df["Number of employees"], errors="coerce")

# Check for missing values in "Number of employees"
missing_employees = df["Number of employees"].isnull().sum()

# Display basic statistics of "Number of employees"
employee_stats = df["Number of employees"].describe()

missing_employees, employee_stats
# Check missing values after fixing
print(f"Missing values in 'Number of employees': {df['Number of employees'].isnull().sum()}")
print(employee_stats)
```

```python
# Group by 'Company_Name / Dénomination_sociale_de_l'entreprise' and 'NAICS / Code_SCIAN' to find average employees
employee_avg = df.groupby(['Company_Name / Dénomination_sociale_de_l\'entreprise', 'NAICS / Code_SCIAN'])['Number of employees'].transform(lambda x: x.fillna(x.mean()))

# Fill missing values in 'Number of employees' column with the grouped mean
df['Number of employees'] = employee_avg

# If some values are still missing (e.g., all rows for a group were NaN), fill with overall median
df['Number of employees'] = df['Number of employees'].fillna(df['Number of employees'].median())
# Check missing values after fixing
print(f"Missing values in 'Number of employees': {df['Number of employees'].isnull().sum()}")

print("Missing values handled successfully!")
```

# 4. Dropping Uneccessary columns

```python
# List of columns to remove
columns_to_remove = [
    "NPRI_ID / No_INRP",
    "Company_Name / Dénomination_sociale_de_l'entreprise",
    "Facility_Name / Installation",
    "Number of employees",
    "NAICS Title / Titre Code_SCIAN",
    "City",
    "CAS_Number / No_CAS"
]

# Remove the specified columns
df_release_cleaned = df.drop(columns=columns_to_remove, errors='ignore')

df_release_cleaned.head()
```

```python
df_release_cleaned.info()
```

```python
# Convert object columns to categorical
object_cols = df_release_cleaned.select_dtypes(include=['object']).columns

# Convert selected columns to categorical
df_release_cleaned[object_cols] = df_release_cleaned[object_cols].astype('category')

# Verify the changes
df_release_cleaned.info()
```

```python
# Step 1: Identify Missing Values
missing_values = df_release_cleaned.isnull().sum()
missing_percentage = (missing_values / len(df_release_cleaned)) * 100

# Create a dataframe to display missing values and their percentage
df_missing = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})

# Step 2: Handle Missing Values
# - Fill numerical missing values with the median of the respective column
# - Fill categorical missing values with the mode (most frequent value)

# Separate numerical and categorical columns
numerical_cols = df_release_cleaned.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df_release_cleaned.select_dtypes(include=['category']).columns

# Fill missing values
df_release_cleaned[numerical_cols] = df_release_cleaned[numerical_cols].fillna(df_release_cleaned[numerical_cols].median())
df_release_cleaned[categorical_cols] = df_release_cleaned[categorical_cols].fillna(df_release_cleaned[categorical_cols].mode().iloc[0])
```

# 6. Handling Outliers in the dataset

```python
# Step 1: Identify Outliers using Interquartile Range (IQR)

# Select only numerical columns for outlier detection
numerical_cols = df_release_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Compute Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df_release_cleaned[numerical_cols].quantile(0.25)
Q3 = df_release_cleaned[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Define outliers as values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
outliers = ((df_release_cleaned[numerical_cols] < (Q1 - 1.5 * IQR)) |
            (df_release_cleaned[numerical_cols] > (Q3 + 1.5 * IQR))).sum()

# Step 2: Replace Outliers with the Median
for col in numerical_cols:
    median_value = df_release_cleaned[col].median()
    lower_bound = Q1[col] - 1.5 * IQR[col]
    upper_bound = Q3[col] + 1.5 * IQR[col]

    # Replace outliers with the median value
    df_release_cleaned[col] = df_release_cleaned[col].apply(
        lambda x: median_value if x < lower_bound or x > upper_bound else x)

# Display the detected outliers before cleaning
# Instead of using ace_tools, use display() from IPython.display
# It's already imported at the top of your notebook
from IPython.display import display
display(outliers.to_frame(name="Outliers Before Cleaning"))


# Verify that outliers have been replaced
outliers_after_cleaning = ((df_release_cleaned[numerical_cols] < (Q1 - 1.5 * IQR)) |
                           (df_release_cleaned[numerical_cols] > (Q3 + 1.5 * IQR))).sum()

# Display the outliers after cleaning using display()
display(outliers_after_cleaning.to_frame(name="Outliers After Cleaning"))
```

**1. Line Graph for Total emissions overtime**

```python
# Aggregate total emissions per year
df['Year'] = pd.to_numeric(df['Reporting_Year / Année'], errors='coerce')
df_yearly = df.groupby("Reporting_Year / Année")["Sum of release to all media (<1tonne)"].sum().reset_index()

#Rename the "Reporting_Year / Année" to "Year"
df_yearly = df_yearly.rename(columns={"Reporting_Year / Année": "Year"})


# Plot emissions over time
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_yearly, x="Year", y="Sum of release to all media (<1tonne)", marker="o", linewidth=2, color="red")
plt.title("Total Emissions Over Time (2000-2020)")
plt.xlabel("Year")
plt.ylabel("Sum of release to all media (<1tonne)")
plt.xticks(rotation=45)
plt.show()
```

**2. Line graph Emissions by Fuel Type (Coal vs. Other Fossil Fuels)**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define NAICS codes for coal-related activities (example)
coal_naics_codes = [
    2121,  # Coal Mining
    221112  # Fossil Fuel Electric Power Generation (which includes coal plants)
]

# Categorize fuel types based on NAICS codes
df['Fuel Type'] = df['NAICS / Code_SCIAN'].apply(
    lambda x: 'Coal' if any(str(code) in str(x) for code in coal_naics_codes) else 'Other Fossil Fuels'
)

# Verify categorization
print(df["Fuel Type"].value_counts())  # Should show counts for 'Coal' and 'Other Fossil Fuels'

# Aggregate emissions by fuel type
df_fuel = df.groupby(["Year", "Fuel Type"])["Sum of release to all media (<1tonne)"].sum().reset_index()

# Plot emissions for coal vs. other fossil fuels
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_fuel, x="Year", y="Sum of release to all media (<1tonne)", hue="Fuel Type", marker="o")
plt.title("Emissions by Fuel Type (Coal vs. Other Fossil Fuels)")
plt.xlabel("Year")
plt.ylabel("Total Emissions (Tonnes)")
plt.legend(title="Fuel Type")
plt.xticks(rotation=45)
plt.show()
```
